```python
#import some libraries
import pandas as pd
import datetime as dt
```

```python
df1=pd.read_csv('https://raw.githubusercontent.com/Raveendra9989152480/weather-prediction/main/data%20sets/train%20(2).csv')
```

```python
df2=pd.read_csv('https://raw.githubusercontent.com/Raveendra9989152480/weather-prediction/main/data%20sets/test.csv')
```

```python
#concatenation of two different tables
data=pd.concat([df1,df2],axis=0)
```

```python
data.reset_index(inplace=True)
```

```python
# to replace the specific values
data['Precip Type'].replace({0:'rain'},inplace=True)
```

```python
# to change the data type 
data['Formatted Date']=pd.to_datetime(data['Formatted Date'],utc=True)
```

```python
data.drop(['index'],axis=1,inplace=True)
```

```python
data['year']=data['Formatted Date'].dt.year
data['month_name']=data['Formatted Date'].dt.month_name()
data['day']=data['Formatted Date'].dt.day
data['month']=data['Formatted Date'].dt.month
```

```python
data.to_csv('Analysis.csv',index=False)
```
